```python
import os
try:
    os.mkdir('reweight_plots')
except:
    pass
from matplotlib import pyplot as plt
import numpy as np
from pylab import *
plt.xkcd()

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.xticks([])
plt.yticks([])

xx = linspace(0,1,100)

fn = lambda xx: 100*xx**6*(xx-0.999)**2+1.5*exp(xx)
plt.plot(xx,fn(xx))
plt.xlabel('Similarity of recommended item to user history')
plt.ylabel('Click through rate')
plt.title('Good recommendations are similar to history,\nbut not too similar')
plt.axis([0, 1, 0, 6.8])

plt.savefig('reweight_plots/rw_aa.png')
```

```python
import scipy.stats.distributions as d
```

```python
xxs = d.beta.rvs(18,1,size=200)
xxs = hstack((xxs,array([0.4,0.66,0.8])))
yys = fn(xxs)
plt.xticks([])
plt.yticks([])
plt.plot(xx,fn(xx))
plt.xlabel('Similarity of recommended item to user history')
plt.ylabel('Click through rate')
plt.title('Historical reco tries to be good,\nit is mostly very similar to history')
plt.plot(xxs,yys,'gx')

plt.axis([0, 1, 0, 6.8])
plt.savefig('reweight_plots/rw_bb.png')
```

```python
xxs = d.beta.rvs(18,1,size=200)
xxs = hstack((xxs,array([0.4,0.66,0.8])))
yys = fn(xxs)
plt.xticks([])
plt.yticks([])
plt.plot(xx,fn(xx))
plt.xlabel('Similarity of recommended item to user history')
plt.ylabel('Click through rate')
plt.title('A linear model underfits.\nThe fit underfit is minimised where data is\nplentiful')
plt.plot(xxs,yys,'gx')
fn(xx)
(4.04423732-4.07752274)/(1-0.98989899)
plt.plot(xx,-3*xx+6.8,'r')
plt.axis([0, 1, 0, 6.8])
plt.savefig('reweight_plots/rw_cc.png')
```

```python
xxs = d.beta.rvs(18,1,size=200)
xxs = hstack((xxs,array([0.4,0.66,0.8])))
yys = fn(xxs)
plt.xticks([])
plt.yticks([])
plt.plot(xx,fn(xx))
plt.xlabel('Similarity of recommended item to user history')
plt.ylabel('Click through rate')
plt.title('When we evaluate new actions we do this uniformly.\nThe model underfits badly where most of the data is.')
plt.plot(xxs,yys,'gx')
xxs2 = d.beta.rvs(1,1,size=20)
yys2 = fn(xxs2)
plt.plot(xxs2,yys2,'kv')

plt.plot(xx,-3*xx+6.8,'r')
plt.axis([0, 1, 0, 6.8])
plt.savefig('reweight_plots/rw_dd.png')
```
